```python
x = list(range(10))
y = x + np.random.rand(10)-0.5

# Calculate the slope and y-intercept of the trendline
fit = np.polyfit(x,y,1)

# Add the trendline
yfit = [n*fit[0] for n in x] + fit[1]
plt.scatter(x,y)
plt.plot(yfit,'black')

plt.show()
```

```python
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)

df['species']=iris['target']

df['species'] = df['species'].apply(lambda x: iris['target_names'][x])

df.head()
```

```python
a= df[df['species'] == iris['target_names'][0]]
b= df[df['species'] == iris['target_names'][1]]
c= df[df['species'] == iris['target_names'][2]]

a.drop('species', axis=1,inplace=True)
b.drop('species', axis=1,inplace=True)
c.drop('species', axis=1,inplace=True)

# Plot the data series
plt.scatter(a.iloc[:,2],a.iloc[:,3], label=iris['target_names'][0])
plt.scatter(b.iloc[:,2],b.iloc[:,3], label=iris['target_names'][1])
plt.scatter(c.iloc[:,2],c.iloc[:,3], label=iris['target_names'][2])

# Label the x- and y-axes
plt.xlabel('Features',size=12,fontweight='semibold')
plt.ylabel('Species',size=12,fontweight='semibold')

# Add grid
plt.grid(linestyle='--',color='#888888',alpha=0.4)

# Add the title and legend
plt.title('Iris Species', size=14, fontweight='semibold')

plt.legend(iris['target_names'])

plt.show()
```

```python
import requests

url = 'https://vignette.wikia.nocookie.net/onepiece/images/c/ca/Wano_Country_Infobox.png/revision/latest?cb=20190626124727'

with open('anim.jpg', 'wb') as f:
    f.write(requests.get(url).content)
```

```python
img = plt.imread('anim.jpg')
plt.imshow(img)
plt.show()
```

```python
# create a copy of the image object from 'plt.imread()' is read-only
imgcopy = img.copy()
imgcopy[img<160] = 0
imgcopy[img>=160] = 222
plt.imshow(imgcopy)
plt.show()
```

```python
# applying pseudocolor  
# default value of colormap is used. 
lum = img[:, :, 0] 
  
# show image 
plt.imshow(lum)
```

```python
lum = img[:, :, 0] 
  
# setting colormap as hot 
plt.imshow(lum, cmap ='hot') 
plt.colorbar() 
```
